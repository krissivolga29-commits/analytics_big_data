rectal temperature - ректальная температура - линейный - в градусах Цельсия. - Повышенная температура может возникнуть из-за инфекции. - температура может быть снижена, когда животное находится в позднем шоке - нормальная температура 37,8. - этот параметр обычно меняется по мере развития проблемы например. может начаться нормально, затем стать повышенным из-за поражение, проходя обратно через нормальный диапазон по мере того, как лошадь впадает в шок

Значения признака rectal temperature имеет диапазон от 35,4 до 40,8.

Скорее всего имеются аномальные значения

```python
# найдем выбросы через межквартильный размах
q1 = df['rectal temperature'].quantile(0.25)
q3 = df['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal temperature'].between(lower_bound, upper_bound, inclusive='both')]
remove_outliers
```

```python
df[~df['rectal temperature'].between(lower_bound, upper_bound, inclusive='both')]
```

```python
print(remove_outliers['rectal temperature'].mean())
print(remove_outliers['rectal temperature'].median())
```

```python
# разница в первоначальном среднем и средним после удаления выбросов почти нулевая. В данном случае удаление выбросов не сыграло
# какой то значимой роли

print(df['rectal temperature'].mean() - remove_outliers['rectal temperature'].mean())
```

```python
# разница в медианах нулевая. Удаление выбросов не
print(df['rectal temperature'].median() - remove_outliers['rectal temperature'].median())
```

pulse - пульс - линейный - частота пульса в ударах в минуту - является отражением состояния сердца: 30-40 – это нормально для взрослых. - редко бывает показатель ниже обычного, хотя у спортивных лошадей может быть ставка 20-25 - животные с болезненными поражениями или страдающие от циркуляторного шока может быть повышенная частота сердечных сокращений.

Значения признака pulse имеет диапазон от 30 до 184.
Имеются аномальные значения

```python
# найдем выбросы через межквартильный размах
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive='both')]
remove_outliers
```

```python
df[~df['pulse'].between(lower_bound, upper_bound, inclusive='both')]
```

```python
# разница в первоначальном среднем и средним после удаления выбросов 1,65. Это связано с большим размахом данных
# (от 30 до 184). В данном случае удаление выбросов не сыграло
# какой то значимой роли, так как разница в медианах равна нулю.

print(df['pulse'].mean() - remove_outliers['pulse'].mean())
```

```python
# разница в медианах нулевая. Удаление выбросов на медиану не повлияло.
print(df['pulse'].median() - remove_outliers['pulse'].median())
```

respiratory rate - частота дыхания - линейный - нормальная ставка от 8 до 10 - полезность сомнительна из-за больших колебаний.

Значения признака respiratory rate имеет диапазон от 8 до 96. Имеются аномальные значения.

```python
# найдем выбросы через межквартильный размах
q1 = df['respiratory rate'].quantile(0.25)
q3 = df['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['respiratory rate'].between(lower_bound, upper_bound, inclusive='both')]
remove_outliers
```

```python
df[~df['respiratory rate'].between(lower_bound, upper_bound, inclusive='both')]
```

```python
# разница в первоначальном среднем и средним после удаления выбросов значительная, 3,7. Это связано с большим размахом данных
# (от 8 до 96). В связи с тем, что полезность данного признака сомнительная, из-за больших колебаний, можно сделать выводы,
# что это естественные значения данных и показатели взяты при разном состоянии животных (в покое, в движении).

print(df['respiratory rate'].mean() - remove_outliers['respiratory rate'].mean())
```

```python
# разница в медианах меньше единицы. Удаление выбросов на медиану значительно не повлияло.
print(df['respiratory rate'].median() - remove_outliers['respiratory rate'].median())
```

#Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

Для количественных данных, если в непрерывной величине есть выбросы, то стоит выбрать для замены медиану, если в непрерывной величине нет выбросов, то стоит выбрать среднее.

```python
#surgery? - категориальный признак, заменим пропуски значением модой, равно 1
#temperature of extremities - категориальный признак, заменим пропуски значением модой, равно 3
#pain - категориальный признак, заменим пропуски значением модой, равно 3
#outcome - категориальный признак, заменим пропуски значением модой, равно 1
#rectal temperature - количественный признак, не имеющий выбросов, заменим медианой
#pulse - количественный признак, имеющий выбросы, заменим средним
#respiratory rate - количественный признак, имеющий выбросы, заменим средним

df_final = df.copy()
df_final['surgery?'] = df_final['surgery?'].fillna(df['surgery?'].mode()[0])
df_final['temperature of extremities'] = df_final['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0])
df_final['pain'] = df_final['pain'].fillna(df['pain'].mode()[0])
df_final['outcome'] = df_final['outcome'].fillna(df['outcome'].mode()[0])
df_final['rectal temperature'] = df_final['rectal temperature'].fillna(df['rectal temperature'].median())
df_final['pulse'] = df_final['pulse'].fillna(df['pulse'].mean())
df_final['respiratory rate'] = df_final['respiratory rate'].fillna(df['respiratory rate'].mean())
```

```python
# пропуски отсутствуют?
(df_final.isna().mean() * 100).round(2)
```
